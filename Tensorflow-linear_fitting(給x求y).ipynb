```python
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

生產玩具data

```python
x = np.linspace(5,15,200)
noise = np.random.randn(200)
y=2*x+2*noise
```

```python
plt.scatter(x,y)
```

# 從keras製造一個線性回歸模組(sequential)

```python
jmda_model=tf.keras.Sequential()
```

2.在模組中插入全連接層dense(第一個1:輸出的數據為幾維,第二個1:輸入的數據為幾維(必須維tuple格式))

```python
jmda_model.add(tf.keras.layers.Dense(1,input_shape=(1,)))
```

param=2(y=ax+b) 線性a,b加起來兩個

```python
jmda_model.summary()
```

3.用compile將model轉譯(optimizer:梯度下降的算法,loss:計算精度(accuracy)的數學方法(mse:mean square error))
*梯度下降:計算loss值與參數的關係，先生成隨機點，再取偏微分取得斜率並沿斜率進行梯度下降，找尋loss值的極小值

```python
jmda_model.compile(optimizer='adam',loss='mse')
```

4.fit數據(epochs:fitting次數)

```python
jmda_model.fit(x,y,epochs=1000)
```

5.觀察做出來的model

```python
y_output=jmda_model.predict(x)
print('y={0}*x+{1}'.format(jmda_model.get_weights()[0][0].round(2),jmda_model.get_weights()[1].round(2)))
```

```python
jmda_model.predict([10])[0][0]
```

```python
plt.scatter(x,y,label='raw data')
plt.plot(x,y_output,'r--',linewidth=4,label='predict data')
plt.legend(loc=2)
```
